```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data1 = np.random.normal(0,0.1,1000)
data2 = np.random.normal(1,0.4,1000)+np.linspace(0,0.1,1000)
data3 = 2 + np.random.random(1000)* np.linspace(1,5,1000)
data4 = np.random.normal(3,0.2,1000)+0.3*np.sin(np.linspace(0,20,1000))
```

```python
data4
```

STACT THESE ARRAYS TO A SINGLE ARRAY 

```python
data = np.vstack([data1,data2,data3,data4]).transpose()
```

```python
data
```

Import it into a Data frame 

```python
df = pd .DataFrame(data, columns=['Data1','Data2','Data3','Data4'])
df.head()
```

Line Plote

```python
df.plot(title="LinePlote")
```

```python
plt.plot(df)
plt.title("LinePlote")
plt.legend(["Data1","Data2","Data3","Data4"])
```

Scatter Plot

```python
# use any of the symbol e.g * _- . ,  try any symbol
df.plot(style=".")
```

Scatter and co relation between two data 

```python
df.plot(kind="scatter",x='Data1', y= 'Data2',
           xlim=(-1.5,1.5),ylim=(0,3))
```

# Histograms

```python
df.plot(
    kind='hist',
    bins=50,
    title='Histogram',
    alpha=0.6,
    figsize=(17,4)
    
)
```

# Cumulative Distribution 

```python
df.plot(
        kind='hist',
        bins= 100,
        title ='Cumulative Distribution',
        normed= True,
        cumulative= True,
        alpha= 0.4,
        figsize=(10,4)
)
```

```python
df.plot(kind="box",
        title='BoxPlot'
       )
```

```python
fig, ax= plt.subplots(2,2, figsize=(16,12))
df.plot(ax=ax[0][0],
       title='linePlot')

df.plot(ax=ax[0][1],
        style='o',
        title="Scatter"
       )

df.plot(ax=ax[1][0],
        kind='hist',
        bins=50,
        title='Historgram'
       )
df.plot(ax=ax[1][1],
        kind='box',
        title='Box Plot'
       )
plt.tight_layout()
```

# Pie Chart

```python
gt = df['Data1'] > 0.1
piecounts=gt.value_counts()
piecounts
```

```python
piecounts.plot(kind='pie',
               figsize=(5,5),
               explode=[0,0.10],
               labels= ['<=0.1','>0.1'],
               autopct='%1.1f%%',
               shadow=True,
               startangle=90,
               fontsize=16
             )
```
