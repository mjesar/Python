```python
df.plot(kind='scatter', x='Temperature', y='WindSpeedMPH',
       title='weather')
plt.plot([10,50], [0,20], color='yellow', linewidth=5)
```

```python
def line(x, w=0, b=0):
    y=w*x+b  #y=mx+c
    return  y
```

```python
def mean_squared_error(y_true, y_pred):
    s=(y_true-y_pred)**2
    return s.mean()
```

```python
X=df[['Temperature']].values
y_true=df[['WindSpeedMPH']].values
```

```python
y_pred=line(X)  #y=mx+c


df.plot(kind='scatter', x='Temperature', y='WindSpeedMPH', 
       title='Weather')

plt.plot([10,50], [0,20], color='yellow', linewidth=3)

plt.plot(X,y_pred, color='red', linewidth=3)
```

```python
mean_squared_error(y_true,y_pred)
```
